```python
# Ableitung mit Moving average
data = polynom_data['with_noise']['y'][1:]
labels = polynom_data['with_noise']['labels'][1:]
x = []
y = []
y.append(data[2])
x.append(np.arange(len(polynom_data['x'])))

average = [2, 4, 8]
levels = len(average)

for n in average:
    #y.append(np.array([np.mean(y[0][i:i+n]) for i in range(0, len(y[0])-n+1)]))
    #x.append(np.array([np.mean(x[0][i:i+n]) for i in range(0, len(x[0])-n+1)]))
    y.append(np.array([np.mean(x) for x in y[-1].reshape(len(y[-1])//2,2)]))
    x.append(np.array([np.mean(x) for x in x[-1].reshape(len(x[-1])//2,2)]))

plt.rcParams["figure.figsize"] = SIZE_FULL
fig, ax_lst = plt.subplots(levels + 3, 1, sharex=True)

ax_lst[0].plot(x[0], y[0], '-o', color='C0', ms=2)
xtick = np.linspace(min(x[0]), max(x[0]), 9)
ax_lst[0].set_xticks(xtick)
ax_lst[0].set_ylabel('Signal')

for i in range(levels + 1):
    dx = x[i][1] - x[i][0]
    ax_lst[i+1].step(x[i][:-1]+dx/2., np.diff(y[i]), 'o', where='mid', color="C1", ms=2)
    ax_lst[i+1].set_ylabel('$\Delta y_{}$'.format(2**i))


ax_lst[levels+2].step(x[-1], y[-1], 'o', where='mid', color="C2", ms=2)
ax_lst[levels+2].set_ylabel('${{y}}_{}'.format(8))

plt.savefig('polynomials_noise_moving_average.pdf')
```
